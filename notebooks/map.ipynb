```python
import gzip
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "../try.json"
with open(path, "r") as f:
    property_data = json.load(f)
```

```python
property_data
```

```python
map_df = pd.DataFrame()


rent = []
name =[]
latitude = []
longitude = []
coordinate =[]
for i in property_data:
    # print(i)
    # print(property_data[i]["name"])
    name.append(property_data[i]["name"])
    rent.append(property_data[i]["cost_text"])
    coordinate.append(property_data[i]["coordinates"])
    latitude.append(property_data[i]["coordinates"][0])
    longitude.append(property_data[i]["coordinates"][1])



map_df['rent'] = rent
map_df['coordinate'] = coordinate
map_df['name'] = name
map_df['longitude'] = longitude
map_df['latitude'] = latitude
```

```python
map_df.head()
```

```python
# due to research
vic_coordinate = [36.9848, 143.3906]
```

```python
import folium
from folium import plugins
```

```python
mapdata = pd.DataFrame()
mapdata['lat'] = map_df['latitude']
mapdata['long'] = map_df['longitude']
mapdata['name'] = map_df['name']
```

```python
vic_map = folium.Map(location=[-37.020100, 144.964600], zoom_start=12)
# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(vic_map)
# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label in zip(mapdata["lat"], mapdata["long"], map_df["name"]):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)
# add incidents to map
vic_map.add_child(incidents)
vic_map
```
